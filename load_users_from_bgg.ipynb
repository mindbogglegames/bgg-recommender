```python
def load_collection( name):
    payload = {'username':name, 'excludesubtype':'boardgameexpansion','stats':1,'rated':1}
    collection_r = requests.get('https://www.boardgamegeek.com/xmlapi2/collection', params=payload)
    
    return collection_r

def load_guild(id, page):
    payload = {'id':id, 'members':1, 'page':page}
    r = requests.get('https://www.boardgamegeek.com/xmlapi2/guild', params=payload)
    
    while(r.status_code != 200):
        r = requests.get('https://www.boardgamegeek.com/xmlapi2/guild', params=payload)
    guild_xml = etree.fromstring(r.text)
    return guild_xml

def parse_col_xml(xml, name):
    collection_xml = etree.fromstring(xml)
    #print(xml)
    col_dict = {}
    for item in collection_xml.iter('item'):
        name = item.find('name')
       
        stats = item.find('stats')
        rating = stats.find('rating')
        
        col_dict[item.get('objectid')] = rating.get('value')
    
    return col_dict
    
def add_members_from_xml(xml):
    collection_r = None
    df = pd.DataFrame()
    for member in xml.iter('member'):
        name = member.get('name')

        if name:
            collection_r = load_collection(name)

            if collection_r.status_code == 200:
                print('adding', name)
                collection_r = load_collection(name)
                col_dict = parse_col_xml(collection_r.text, name)
                series = pd.Series(col_dict, name=name, dtype='float')
                df = df.append(series)
            else:
                tries = 0
                while(collection_r.status_code != 200 and tries < 20):
                    tries+=1
                    time.sleep(2) 
                    collection_r = load_collection(name)
                
                if(collection_r.status_code == 200):   
                    print('adding', name)
                    col_dict = parse_col_xml(collection_r.text, name)
                    series = pd.Series(col_dict, name=name, dtype='float')
                    df = df.append(series)
                else:
                    print('skipping', name)
    return df

def print_members_from_collection(xml):
    for member in xml.iter('member'):
        name = member.get('name')
        print(name)
```
